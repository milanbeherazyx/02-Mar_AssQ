Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular open-source data visualization library in Python that is used to create static, animated, and interactive visualizations in Python. It is a 2D plotting library that provides a wide range of tools for creating high-quality graphs, charts, and figures. Matplotlib is widely used in the scientific community and data analysis to create publication-quality graphics, visualizations, and plots.

Matplotlib is used to visualize data and communicate insights to others in a clear and concise way. It provides a wide range of visualization techniques such as line plots, scatter plots, bar plots, histograms, heat maps, and many more. The library is highly customizable and provides a lot of control over the look and feel of the visualizations.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot - Line plots are used to visualize the relationship between two variables by plotting a line that connects the data points.

Scatter Plot - Scatter plots are used to visualize the relationship between two variables by plotting points on a 2D plane, where the position of each point represents the values of the two variables.

Bar Plot - Bar plots are used to compare different categories of data by plotting bars of different heights representing the data values.

Histogram - Histograms are used to visualize the distribution of data by grouping data into bins and plotting the number of data points that fall into each bin.

Pie Chart - Pie charts are used to visualize the relative sizes of different categories of data by dividing a circle into sectors, where each sector represents a category and the size of the sector represents the proportion of data in that category.






```python
#line plot

# Create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin Wave')
plt.show()
```

```python
# Scatter plot


# Create some data
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

# Plot the data
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random Data')
plt.show()
```

```python
# Bar plot


# Create some data
x = ['A', 'B', 'C', 'D', 'E']
y = [20, 35, 30, 25, 40]

# Plot the data
plt.bar(x, y)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart')
plt.show()
```

```python
# Histogram


# Create some data
x = np.random.normal(0, 1, size=1000)

# Plot the data
plt.hist(x, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Pie chart

# Create some data
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [20, 35, 30, 25, 40]

# Plot the data
plt.pie(sizes, labels=labels)
plt.title('Pie Chart')
plt.show()
```

A scatter plot is a type of plot that is used to visualize the relationship between two variables. It displays the data as a collection of points, with each point representing the value of one variable relative to the other. Scatter plots are particularly useful for identifying patterns or trends in data and for detecting outliers or anomalies.

Here is an example of how to generate a scatter plot using the given code:

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function is used to create multiple plots within a single figure. It allows you to arrange multiple plots in a grid-like pattern, where each plot represents a different subset of the data or a different analysis.

Here is an example of how to use the subplot() function to create four line plots using the given data:

```python
# Define the data for each line plot
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots arranged in a 2x2 grid
fig, axes = plt.subplots(nrows=2, ncols=2)

# Plot each line plot in a different subplot
axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('Line Plot 1')

axes[0, 1].plot(x2, y2)
axes[0, 1].set_title('Line Plot 2')

axes[1, 0].plot(x3, y3)
axes[1, 0].set_title('Line Plot 3')

axes[1, 1].plot(x4, y4)
axes[1, 1].set_title('Line Plot 4')

# Add a title to the entire figure
fig.suptitle('Four Line Plots')

# Display the plot
plt.show()
```

In this example, we first define the data for each line plot using the np.array() function. We then create a figure with four subplots arranged in a 2x2 grid using the plt.subplots() function, which returns both the figure and an array of axes objects representing each subplot. We then plot each line plot in a different subplot using the plot() function and set the title for each subplot using the set_title() function. Finally, we add a title to the entire figure using the fig.suptitle() function and display the plot using the plt.show() function.

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a way to visualize data using rectangular bars, where the length of each bar is proportional to the value of the corresponding data point. Bar plots are used to compare different categories or groups of data, and they can be used to show trends over time, differences between groups, or rankings of categories.

Here's how to create a bar plot and a horizontal bar plot using the given data:

```python
# Define the data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.bar(company, profit)
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.show()

# Create a horizontal bar plot
plt.barh(company, profit)
plt.title('Company Profits')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.show()
```

In the first plot, we create a vertical bar plot using the plt.bar() function and pass in the company array as the x-axis values and the profit array as the y-axis values. We then add a title, x-label, and y-label using the plt.title(), plt.xlabel(), and plt.ylabel() functions, respectively. Finally, we display the plot using the plt.show() function.

In the second plot, we create a horizontal bar plot using the plt.barh() function and pass in the company array as the y-axis values and the profit array as the x-axis values. We then add a title, x-label, and y-label using the same functions as before. Finally, we display the plot using the plt.show() function.

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a way to visualize the distribution of a dataset. It displays the median, quartiles, and outliers of the data in a compact way. The box in the middle of the plot represents the interquartile range (IQR), which contains the middle 50% of the data. The whiskers extend from the box to show the range of the data, excluding outliers, which are shown as individual points.

Box plots are used to identify the range and distribution of a dataset, as well as to identify any outliers. They are particularly useful when comparing multiple datasets, as they allow for easy visual comparison of their distributions.

Here's how to create a box plot using the given data:

```python
# Define the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a list
data = [box1, box2]

# Create a box plot
plt.boxplot(data)
plt.title('Box Plot of Box1 and Box2')
plt.xlabel('Box')
plt.ylabel('Value')
plt.xticks([1, 2], ['Box1', 'Box2'])
plt.show()
```

In this example, we first define the two datasets as box1 and box2 using the np.random.normal() function to generate random data with a normal distribution. We then combine the data into a list called data, which we pass to the plt.boxplot() function to create the box plot.

We add a title, x-label, and y-label using the plt.title(), plt.xlabel(), and plt.ylabel() functions, respectively. We also set the x-tick labels using the plt.xticks() function to display 'Box1' and 'Box2' instead of the default values of '1' and '2'. Finally, we display the plot using the plt.show() function.
